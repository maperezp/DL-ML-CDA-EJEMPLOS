Función para cálculo de tiempo, ideal para procesos extensos:

```python
def timer(start_time=None):
  if not start_time:
    start_time = datetime.now()
    return start_time
  elif start_time:
    thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
    tmin, tsec = divmod(temp_sec, 60)
    print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

Para visualizar la matriz de confusión:

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues,size=(4, 4)):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    fig, ax = plt.subplots(figsize=size)
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

Para visualizar la curva ROC:

```python
def display_multiple_models_roc_curve(models, X, y):
  # Below for loop iterates through your models list
  for m in models:
    model = m['model'] # select the model
    y_pred=model.predict(X) # predict the test data
    # Compute False postive rate, and True positive rate
    fpr, tpr, thresholds = roc_curve(y, model.predict_proba(X)[:,1])
    # Calculate Area under the curve to display on the plot
    auc = roc_auc_score(y,y_pred)
    # Now, plot the computed values
    plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % (m['label'], auc))
  # Custom settings for the plot
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])
  plt.xlabel('1-Specificity(False Positive Rate)')
  plt.ylabel('Sensitivity(True Positive Rate)')
  plt.title('Receiver Operating Characteristic')
  plt.legend(loc="lower right")
  plt.show()   # Display
```

## 2.1 Preprocess

```python
# Variables a eliminar
drop_columns = ['High Temperature (°C)','Low Temperature (°C)', 'Low Oxygen Alert', 'Thermal Risk Index']
```

```python
# Preprocess
def preprocess(df, drop_fields):
  df = df.drop_duplicates() # Eliminamos duplicados
  df = df.drop(drop_columns, axis = 1) # Eliminamos columnas que no vamos a usar
  # df['Oxygenation Automatic'] = le.fit_transform(df['Oxygenation Automatic']) Convertir la columna Oxygenation Automatic en binaria
  X_data, y_variable = df.drop([target_feature], axis=1), df[target_feature]
  y_variable = y_variable.replace({'Stable': 1, 'At Risk': 0}).astype(int)
  return X_data, y_variable
```

## 2.2 Pipeline

```python
def convertir_yes_no(df, columna):
    df[columna] = df[columna].map({'Yes': 1, 'No': 0})
    return df

def round_to_int(X):
    return np.round(X).astype(int)
```

```python
# Crear el transformer con nombre
num_transformer = Pipeline(steps=[
    ('round_to_integer_disease', FunctionTransformer(round_to_int, feature_names_out="one-to-one"))
])

cat_transformer = Pipeline(steps=[
    ("label_encoding_Disease", FunctionTransformer(convertir_yes_no, kw_args={'columna': 'Oxygenation Automatic'}, feature_names_out="one-to-one"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("disease", num_transformer, ["Disease Occurrence"]),
        ("oxy_aut", cat_transformer, ["Oxygenation Automatic"]),
        ("month", OneHotEncoder(handle_unknown="ignore"), ["Month"]),
        ("others", "passthrough", ["Average Fish Weight (g)", "Survival Rate (%)", "Temperature (°C)", "Dissolved Oxygen (mg/L)", "pH", "Turbidity (NTU)", "Oxygenation Interventions", "Corrective Interventions"])
    ]
)
```

```python
pipe = Pipeline(steps=[
    ('column_transformer', preprocessor)
])

pipe
```

```python
pipe.fit_transform(X_train)
```

```python
pipe.get_feature_names_out()
```

A continuación, encontrarás el uso de los parámetros más relevantes para DecisionTreeClassifier.

* `criterion`: Este parámetro determina la función para medir la calidad de una partición. Los criterios admitidos incluyen "gini" para la impureza de Gini, y "log_loss" y "entropy", que representan la ganancia de información de Shannon. El valor por defecto es "gini".
* `splitter`: Este parámetro define la estrategia utilizada para elegir la partición en cada nodo. Las estrategias admitidas son "best" para elegir la mejor partición y "random" para elegir la mejor partición aleatoria. El valor por defecto es "best".
* `max_depth`: Este parámetro especifica la profundidad máxima del árbol. Si se establece en None, los nodos se expanden hasta que todas las hojas son puras o contienen menos de `min_samples_split` muestras. El valor por defecto es "None".
* `min_samples_split`: Este parámetro establece el número mínimo de muestras requeridas para dividir un nodo interno. Puede ser un entero o un flotante, donde un flotante representa una fracción, y el número mínimo de muestras para cada partición se calcula como $ceil(min\_samples\_split \times n\_samples)$. El valor por defecto es "2".
* `min_samples_leaf`: Este parámetro determina el número mínimo de muestras requeridas para estar en un nodo hoja. Similar a `min_samples_split`, puede ser un entero o un flotante. El valor por defecto es "1".
* `max_features`: Este parámetro especifica el número de características a considerar cuando se busca la mejor partición. Puede ser un entero, un flotante o una de las cadenas "sqrt" o "log2". El valor por defecto es "None".

En la variable `estimators_bm_tree` vamos a configurar los siguientes valores:

* `transform` que contiene el preprocesamiento de la información.
* `classifier` con la información del estimador, que en este caso corresponderá a `DecisionTreeClassifier`, dónde especificaremos una profundidad máxima del arbol igual a 3 y la semilla de reproducibilidad `random_state`. El resto de parámetros del clasificados se dejan por defecto.

```python
%%time
estimators_bm_tree = [
        ('transform', preprocessor),
        ('classifier', DecisionTreeClassifier( max_depth=3, random_state=my_seed))
]

pipe_bm_tree = Pipeline(estimators_bm_tree)
```
